```python
# Clean the BPD data set so that it's a bit easier to use

lines = []
for line in open('BPD_Arrests.csv').readlines():
    if line[-1] == ',': continue   # if it doesn't have a geo location

    # some quotes that are annoying
    nl = line.replace('"(', '').replace(')"','').replace(', ', ',') 

    # if the number of tokens isn't 16, i'm not going to worry about it
    if len(nl.split(',')) != 16: continue
        
    lines.append(nl)
    
# write it back out for later use
open('BPD_Arrests.clean.csv','w').write(''.join(lines))
```

```python
# cluster events
import time
import datetime
from sklearn.preprocessing import scale

SCALE = 2500000000.

def convert_time(t):
    return float(int(time.mktime(time.strptime(t, "%m %d %H:%M:%S %Y"))) / SCALE)

def time2str(t):
    return str(datetime.datetime.fromtimestamp(t * SCALE) ).split('.')[0]
```

```python
cluster_data = pandas.DataFrame(\
                        [ [convert_time('04 ' + b['created_at'][8:19] + ' 2015') ] \
                         + b['geo']['coordinates'] for b in bmoregeo ], \
                        columns=['Time', 'Latitude', 'Longitude']).drop_duplicates()
```

```python
from sklearn.cluster import KMeans
km = KMeans(init='k-means++', n_clusters=90, n_init=5)
kmf = km.fit(cluster_data)

centers = pandas.DataFrame(kmf.cluster_centers_, columns=['Time', 'Latitude', 'Longitude'])
centers['STime'] = centers['Time'].map(time2str)

centers = centers.join(\
    pandas.DataFrame(pandas.Series(kmf.predict(cluster_data)).value_counts(), columns=['Counts'])   )

centers = centers.sort(['Counts'])

#
```
